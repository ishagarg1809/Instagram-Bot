```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException

driver = webdriver.Chrome(executable_path=r"C:\Users\Rohan Garg\Desktop\Isha Master File\chrome driver\chromedriver.exe")
driver
```

```python
#Ques 1
def login_account(username, password) :
    driver.get('https://www.instagram.com/?hl=en')
    
    time.sleep(2)
    
    #To fill in username : first find it and then enter values
    user = driver.find_element_by_name('username')
    user.send_keys(username)

    #Finding and then filling the password tab
    passw = driver.find_element_by_name('password')
    passw.send_keys(password)

    #Finding submit button
    button = driver.find_element_by_xpath('//button/div')
    button.click()

    time.sleep(5)

    pop_up = driver.find_elements_by_xpath('//div[contains(@class, "mt3GC")]/button')[-1]
    pop_up.click()
    
login_account('SAMPLE USERNAME', 'SAMPLE PASSWORD')
```

```python
#Ques 4, 2)
def follow(username) :
    driver.get('https://www.instagram.com/'+ username+'/')
    try :
        driver.find_element_by_xpath('//div[contains(@class, "_862NM")]/div/button')
        print('Already follow this account')
    
    except NoSuchElementException :
        follow = driver.find_element_by_xpath('//span[contains(@class, "vBF20 _1OSdk")]/button')
        follow.click()
        return
    
follow('sodelhi')
```

```python
#Ques 4, 3)
def unfollow(username) :
    driver.get('https://www.instagram.com/'+ username+'/')
    try :
        driver.find_element_by_xpath('//div[contains(@class, "_862NM")]/div/button')
        
        driver.find_element_by_xpath('//span[contains(@class, "vBF20 _1OSdk")]/button').click() #To click on unfollow
        driver.find_element_by_xpath('//div[contains(@class, "mt3GC")]/button').click() #To check the unfollow dialogue box
    
    except NoSuchElementException :
        print('Already unfollowed this account')
        return
    
unfollow('sodelhi')
```

```python
#Ques 5, 1)

def like(username) :
    driver.get('https://www.instagram.com/'+ username+'/')
    driver.find_element_by_xpath('//div[contains(@class, "eLAPa")]').click()   ##To open the first post
    time.sleep(1)
    for i in range(30) :
        if driver.find_elements_by_class_name('_8-yf5 ')[8].get_attribute('aria-label')=='Like':
            driver.find_element_by_xpath('//span[contains(@class, "fr66n")]/button').click()
            
        else :
            print('Already liked it')
         
        driver.find_elements_by_xpath('//div[contains(@class, "DdSX2")]/a')[-1].click()  ##To click on next post
        time.sleep(2)
            
            
like('dilsefoodie')
```

```python
#Ques 6, 1)
def followers_list(username) :
    driver.get('https://www.instagram.com/'+ username+'/')
    time.sleep(1)
    driver.execute_script('document.documentElement.scrollTop=0;')    #To scroll to the top of the page
    
    n = driver.find_element_by_xpath('//a[contains(@class, "-nal3 ")]/span')
    n.click()      ##To click on followers
    
    followers_panel = driver.find_element_by_xpath('//div[@class = "isgrP"]')
    len_followers = len(followers_panel.find_elements_by_css_selector('li'))
    
    #To scroll down the dialog box
    while len_followers < 500 :
        new_height = driver.execute_script("return arguments[0].scrollHeight",followers_panel)
        driver.execute_script("arguments[0].scrollTo(0, arguments[1]);",followers_panel,new_height)
        len_followers = len(followers_panel.find_elements_by_css_selector('li'))
    
    follow_list = driver.find_elements_by_class_name('uu6c_')
    
    follower=set()
    for i in range(500) :
        user = follow_list[i].find_element_by_class_name('_0imsa').text
        follower.add(user)
    
    print(len(follower))
        
followers_list('foodtalkindia')
followers_list('sodelhi')
```

```python
#Ques 6, 2)

def followers_dont_follow_back(username) :
    driver.get('https://www.instagram.com/'+ username+'/')
    
    try:
        driver.find_element_by_class_name('tc8A9').click()  #To click on mutual friends

    except NoSuchElementException :
        print('No mutual followers')
        return
        
    time.sleep(5)
    driver.find_element_by_class_name('ZIAjV').click() # to see all followers
    
    time.sleep(1)
    followers_panel = driver.find_element_by_xpath('//div[@class = "isgrP"]')
    
    len_followers = len(followers_panel.find_elements_by_css_selector('li'))
    mutual_followers = len(followers_panel.find_elements_by_class_name('_8A5w5'))
    
    #To scroll down the dialog box
    while len_followers < mutual_followers :
        new_height = driver.execute_script("return arguments[0].scrollHeight",followers_panel)
        driver.execute_script("arguments[0].scrollTo(0, arguments[1]);",followers_panel,new_height)
        len_followers = len(followers_panel.find_elements_by_css_selector('li'))
        
    #Extracting all mutual followers
    fol = set()

    for i in driver.find_elements_by_css_selector('li') :
        try :
            i.find_element_by_class_name('_8A5w5')
            username = i.find_element_by_class_name('_0imsa').text
            fol.add(username)

        except :
            EC.NoSuchElementException
            
    time.sleep(5)
    driver.find_elements_by_class_name('wpO6b ')[1].click() #To remove the dialogue box
    
    #opening personal profiles
    profile = driver.find_element_by_xpath('//a[contains(@class, "_2dbep qNELH kIKUG")]').get_attribute('href')
    driver.get(profile)
    
    n = driver.find_element_by_xpath('//a[contains(@class, "-nal3 ")]/span')  #To see followers
    num_followers = int(float(n.text))
    n.click()   #Open followers dialogue box
    
    time.sleep(1)
    followers_panel = driver.find_element_by_xpath('//div[@class = "isgrP"]')
    len_followers = len(followers_panel.find_elements_by_css_selector('li'))
    

    #To scroll down the dialog box
    while len_followers < 500 :
        new_height = driver.execute_script("return arguments[0].scrollHeight",followers_panel)
        driver.execute_script("arguments[0].scrollTo(0, arguments[1]);",followers_panel,new_height)
        len_followers = len(followers_panel.find_elements_by_css_selector('li'))
        

    #Finding followers who I follow back
    following_me_back = set()

    for i in driver.find_elements_by_class_name('uu6c_') :
        try :
            i.find_element_by_class_name('_8A5w5')
            username = i.find_element_by_class_name('_0imsa').text
            following_me_back.add(username)

        except :
            EC.NoSuchElementException

    time.sleep(10)
    
    #Printing followers of the page that don't follow me back
    for i in fol :
        if i not in following_me_back :
            print(i)
        
    driver.find_element_by_xpath('//div[contains(@class, "WaOAr")]/button').click()   ##To cross the dialogue box
    

followers_dont_follow_back('foodtalkindia')
```

```python
#Ques 7

def story(username) :
    driver.get('https://www.instagram.com/'+ username+'/')
    
    try :
        driver.find_element_by_xpath('//div[contains(@class, "XjzKX")]/div')
        time.sleep(1)
        
        if int(float(driver.find_element_by_xpath('//div[contains(@class, "XjzKX")]/div/canvas').get_attribute('height'))) == 208 :
            print("You have already seen the user's story")
            
        else :
            driver.find_element_by_xpath('//div[contains(@class, "XjzKX")]/div').click()
    
    except NoSuchElementException :
        print('The user has no story')
    
story('coding.ninjas')
```
